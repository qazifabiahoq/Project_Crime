```python
#setup
import csv
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f_oneway
from scipy.stats import chi2_contingency

filepath=os.path.join("Resources", "Crime_Data_from_2020_to_Present.csv")
crimedata=pd.read_csv(filepath)
crimedata
```

```python
#Statistical Test 1: Perform an anova test to compare crime counts between different areas. Identify if there are significant differences.

#making a dictonary to visualize each area number with its name
temp=int()
areadic={}
for i in range(len(crimedata['AREA'])):
    if crimedata['AREA'][i]!=temp:
        temp=int(crimedata['AREA'][i])
        areadic[temp]=crimedata['AREA NAME'][i]
#sorting numerically
areadic2=dict(sorted(areadic.items()))
areadic2
```

```python
#making a dictonary to visualize each crime code and its description
temp2=int()
crimecodedic={}
for i in range(len(crimedata['Crm Cd'])):
    if crimedata['Crm Cd'][i]!=temp2:
        temp2=int(crimedata['Crm Cd'][i])
        crimecodedic[temp2]=crimedata['Crm Cd Desc'][i]
        
#sorting numerically
ccd2=dict(sorted(crimecodedic.items()))
ccd2
```

```python
#ANOVA test:

#making the Crm_Cd Column numeric
crimedata['Crm Cd']=pd.to_numeric(crimedata['Crm Cd'], errors='coerce')

#taking out any entries that are invalid
crimedata=crimedata.dropna(subset=['Crm Cd'])

#calculating crime counts and grouping them by area
groupeddata=crimedata.groupby('AREA')['Crm Cd'].count()

#making my varaibles for the 'f_oneway' function
crimecountperarea=[groupeddata[area] for area in groupeddata.index]
areas=groupeddata.index

#running anova test, alpha = 0.05
#H0: mean crime rate is the same in all areas
#H0: at least one area has a significantly different crime rate
f_statistic, p_value = f_oneway(areas, crimecountperarea)
print("F-statistic:", f_statistic)
print("p-value:", p_value)
```

```python
#with a p-value of ~2.53, we fail to reject the null hypothesis
#Therefore, there is not enough evidence to conclude that there exist significant differences in crime count between the 21 areas
```

```python
#Statistical Test 2.1: Correlation test for Monthly crime trends over the years (are crime counts consistent during specific months over different years)

#making the DATE_OCC columns datetime format
crimedata['DATE OCC']=pd.to_datetime(crimedata['DATE OCC'])

#making additional columns for month and year
crimedata['month']=crimedata['DATE OCC'].dt.month
crimedata['year']=crimedata['DATE OCC'].dt.year

#making a pivot table to count the number of crimes for each month and year
pivottable=crimedata.pivot_table(index='month', columns='year', values='DR NO', aggfunc='count')
pivottable
```

```python
#producing the correlation matrix
correlationmatrix=pivottable.corr()
correlationmatrix
```

```python
#weak = 0 - 0.3
#med = 0.3 - 0.7
#strong = 0.7 - 1
#Therefore, weak to medium correlations across the board, meaning there is no particular month thats sees a consistent trend
```

```python
#Statistical Test 2.2: Checking if any particular month sees an increase in crime:

#making a dictonary to store the total crime counts for each month over the 4 year span
temp3=int(i)
monthlycountdic={}
for i in range(len(crimedata['month'])):
    temp3=int(crimedata['month'][i])
    if temp3 not in monthlycountdic:
        monthlycountdic[temp3]=1
    else:
        monthlycountdic[temp3]+=1
monthlycountdic=dict(sorted(monthlycountdic.items()))   
   
#running anova test:
#H0: the mean crime rates across all months are the same
#H1: at least one month has a significantly different crime rate
monthlist=list(monthlycountdic.keys())
monthlycountlist=list(monthlycountdic.values())
f_statistic, p_value = f_oneway(monthlist, monthlycountlist)
print("F-statistic:", f_statistic)
print("p-value:", p_value)
```

```python
#with a p-value of ~3.04, we fail to reject the null hypothesis
#Therefore, there is not enough evidence to conclude that there exist significant differences in crime count between the months
```

```python
#Statistical Test 3.1: Chi-square test to analyze the association between crime types and other categorical variables (Area).

#make contigency table
contingencytable = pd.crosstab(crimedata['Crm Cd Desc'], crimedata['AREA'])
contingencytable
```

```python
#the chi-squared test:
#H0: there is no significant association between Crm Cd Desc and AREA
#H1: there is a significant association between Crm Cd Desc and AREA
chi2_stat, p_value, dof, expected=chi2_contingency(contingencytable)

print("\nChi-squared statistic:", chi2_stat)
print("P-value:", p_value)
print("Degrees of freedom:", dof)
print("expected frequencies table:")
print(pd.DataFrame(expected, index=contingencytable.index, columns=contingencytable.columns))
```

```python
#with this extremely low p-value, we reject the null hypothesis
#Therefore, we can conclude there is an association between crime type and area
```

```python
#Statistical Test 3.2: Vehicle - Stolen seems to be the most common offence; test this variable vs. area to see if contribution to this stat is even or specifically a few areas
#H0: even contribution to vehicle theft from all areas
#HA: one area contributes significantly to vehicle theft

#pulling out just the VEHICLE - STOLEN data from the table above
stolenvehiclesrow=contingencytable.loc['VEHICLE - STOLEN'].values
stolenvehiclesrow
```

```python
#running anova test:
f_statistic, p_value= f_oneway(areas, stolenvehiclesrow)
print("F-statistic:", f_statistic)
print("p-value:", p_value)
```

```python
#with a p-value of ~1.45, we fail to reject the null
#Therefore, there is not enough evidence to conclude that one area significantly contributes to vehicle theft
```

```python
#Statistical Test 4: T-tests or other appropriate tests for comparisons between different demographic groups.

#making a dictionary to store the totals for each age:
temp4=int(i)
agecount={}
for i in range(len(crimedata['Vict Age'])):
    temp4=crimedata['Vict Age'][i]
    if crimedata['Vict Age'][i] not in agecount:
        agecount[temp4]=1
    else:
        agecount[temp4]+=1
agecount=dict(sorted(agecount.items()))
agecount
```
